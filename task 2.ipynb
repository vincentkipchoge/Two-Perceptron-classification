## Import Libraries

```python
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pandas as pd
```

## Load Data Using Pandas

```python
df= pd.read_csv('emails.csv')
```

```python
df = df.drop(columns=['Email No.'])
 
```

```python
df
```

```python
y= df['Prediction']
X= df.drop(columns = ['Prediction'])
```

```python
X
```

```python
y
```

```python
train_x,test_x, train_y,test_y = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Standardize the featuress
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
train_x = scaler.fit_transform(train_x)
test_x = scaler.transform(test_x)

```

## Define Parameters

```python
#Using sigmoid as activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
#derivative of the sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)
# Initialize random weights
np.random.seed(42)
weights = 2 * np.random.random((X.shape[1], 1)) - 1
# Set the number of epochs
epochs = 1000

#the learning rate
lr = 0.1

```

## Creating and Training Single Layer Perceptron (SLP)

```python
#Train the neural network
for epoch in range(epochs):
    # Forward pass
    input_layer = train_x
    output = sigmoid(np.dot(input_layer, weights))

    # Calculate the error
    error = train_y.values.reshape(-1, 1) - output

    # Backpropagation
    adjustment = lr * np.dot(input_layer.T, error * sigmoid_derivative(output))

    # Update weights
    weights += adjustment
```

## Testing

```python
# Testing the neural network
test_output = sigmoid(np.dot(test_x, weights))
# Convert the output to binary (0 or 1)
predictions = np.round(test_output)

accuracy = np.mean(predictions == test_y.values.reshape(-1, 1))
accuracy = np.mean(predictions.flatten() == test_y.values)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
#Calculate precision
predictions_flat = predictions.flatten()
test_y_flat = test_y.values.flatten()
precision = precision_score(test_y_flat, predictions_flat)
# Calculate recall
recall = recall_score(test_y_flat, predictions_flat)
# Calculate F1 score
f1 = f1_score(test_y_flat, predictions_flat)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.metrics import confusion_matrix
# Convert predictions and true labels to 1D arrays
predictions_flat = predictions.flatten()
test_y_flat = test_y.values

# Calculate confusion matrix
conf_matrix = confusion_matrix(test_y_flat, predictions_flat)
print("Confusion Matrix:")
print(conf_matrix)
```
